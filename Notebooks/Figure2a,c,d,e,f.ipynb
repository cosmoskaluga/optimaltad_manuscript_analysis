## Figure 2a

```python
prefix = "/Users/dmitriismirnov/Dropbox/Projects/optimalTAD/Results/Lamina_2019/ChIPseq/No_Imputation_new/"
```

```python
stair = pd.read_csv(prefix + 'stair.csv')
```

```python
stair.columns = ['Distance', 'LacZ 1', 'LacZ 2', 'Lamin 1', 'Lamin 2']
```

```python
df1 = stair[['Distance','LacZ 1']]
df2 = stair[['Distance','LacZ 2']]
df3 = stair[['Distance','Lamin 1']]
df4 = stair[['Distance','Lamin 2']]

df1.columns = ['Distance', 'Value']
df2.columns = ['Distance', 'Value']
df3.columns = ['Distance', 'Value']
df4.columns = ['Distance', 'Value']

poly = np.polyfit(df1.Distance.values, df1.Value.values, 3)
poly_y = np.poly1d(poly)(df1.Distance.values)
df1['Value(smoothed)'] = poly_y

poly = np.polyfit(df2.Distance.values, df2.Value.values, 3)
poly_y = np.poly1d(poly)(df2.Distance.values)
df2['Value(smoothed)'] = poly_y

poly = np.polyfit(df3.Distance.values, df3.Value.values, 3)
poly_y = np.poly1d(poly)(df3.Distance.values)
df3['Value(smoothed)'] = poly_y

poly = np.polyfit(df4.Distance.values, df4.Value.values, 3)
poly_y = np.poly1d(poly)(df4.Distance.values)
df4['Value(smoothed)'] = poly_y

df1['Replicate'] = 'Control 1'
df2['Replicate'] = 'Control 2'
df3['Replicate'] = 'Lam-KD 1'
df4['Replicate'] = 'Lam-KD 2'

df_stairs = pd.concat([df1, df2, df3, df4])
```

```python
#df_stairs[['Distance']] = df_stairs[['Distance']]*20 
```

```python
sns.set_theme(style="ticks")

fig, ax = plt.subplots(figsize=(7, 7))
sns.set_context("paper", font_scale=2.5)

my_pal = {"Control 1": "#2b8cbe", "Control 2": "#a6bddb", "Lam-KD 1": "#de2d26", "Lam-KD 2": "#e5a0a0"}
sns.lineplot(data = df_stairs, x = "Distance", 
             y = "Value(smoothed)", hue = "Replicate", 
            linewidth = 3, palette = my_pal)

sns.despine(offset=5, trim=False)
plt.ylabel('Median z-score of acetylation values', fontsize = 25, labelpad = 2)
plt.xlabel('Distance to TAD boundary, kb', fontsize = 25, labelpad = 70)

plt.yticks(np.arange(-0.6, 1.01, 0.2))
plt.setp(ax.get_legend().get_texts(), fontsize='19') # for legend text
plt.setp(ax.get_legend().get_title(), fontsize='23') # for legend title

ax.set_xticks([-4, -3, -2, -1, 0, 1, 2, 3, 4])
ax.set_xticklabels([-80, -60, -40, -20, 0, 20, 40, 60, 80])

plt.grid()
```

## Figure 2f

```python
total_Lacz1_Lacz2 = 0
total_Lacz1_Lamina1 = 0
total_Lacz1_Lamina2 = 0

len_Lacz1 = 0
len_Lacz2 = 0
len_Lamina1 = 0
len_Lamina2 = 0

for name in chrnames:
    d_Lacz1 = tads_Lacz1.loc[tads_Lacz1.Chr == name]
    d_Lacz2 = tads_Lacz2.loc[tads_Lacz2.Chr == name]
    d_Lamina1 = tads_Lamin1.loc[tads_Lamin1.Chr == name]
    d_Lamina2 = tads_Lamin2.loc[tads_Lamin2.Chr == name]
    
    all_Lacz1 = np.append(d_Lacz1.Start.values, d_Lacz1.End.values)
    all_Lacz2 = np.append(d_Lacz2.Start.values, d_Lacz2.End.values)
    all_Lamina1 = np.append(d_Lamina1.Start.values, d_Lamina1.End.values)
    all_Lamina2 = np.append(d_Lamina2.Start.values, d_Lamina2.End.values)
    
    all_Lacz1 = np.unique(all_Lacz1)
    all_Lacz2 = np.unique(all_Lacz2)
    all_Lamina1 = np.unique(all_Lamina1)
    all_Lamina2 = np.unique(all_Lamina2)
    
    len_Lacz1 += len(all_Lacz1)
    len_Lacz2 += len(all_Lacz2)
    len_Lamina1 += len(all_Lamina1)
    len_Lamina2 += len(all_Lamina2)
    
    Lacz1_Lacz2 = 0
    for b1 in all_Lacz1:
        for b2 in all_Lacz2:
            if b1 == b2 or b1+20000 == b2 or b1-20000 == b2:
                Lacz1_Lacz2+=1
                all_Lacz2 = np.delete(all_Lacz2, np.where(all_Lacz2 == b2))
                break
    
    Lacz1_Lamina1 = 0
    for b1 in all_Lacz1:
        for b2 in all_Lamina1:
            if b1 == b2 or b1+20000 == b2 or b1-20000 == b2:
                Lacz1_Lamina1+=1
                all_Lamina1 = np.delete(all_Lamina1, np.where(all_Lamina1 == b2))
                break
                
    Lacz1_Lamina2 = 0
    for b1 in all_Lacz1:
        for b2 in all_Lamina2:
            if b1 == b2 or b1+20000 == b2 or b1-20000 == b2:
                Lacz1_Lamina2+=1
                all_Lamina2 = np.delete(all_Lamina2, np.where(all_Lamina2 == b2))
                break
    
    total_Lacz1_Lacz2 += Lacz1_Lacz2
    total_Lacz1_Lamina1 += Lacz1_Lamina1
    total_Lacz1_Lamina2 += Lacz1_Lamina2
    
print('Total number of borders, Lacz1: ', len_Lacz1)
print('Total number of borders, Lacz2: ', len_Lacz2)
print('Total number of borders, Lamina1: ', len_Lamina1)
print('Total number of borders, Lamina2: ', len_Lamina2)
print(' ')
print('Shared borders, Lacz1 and Lacz2: ', total_Lacz1_Lacz2)
print('Shared borders, Lacz1 and Lamina1: ', total_Lacz1_Lamina1)
print('Shared borders, Lacz1 and Lamina2: ', total_Lacz1_Lamina2)
```

```python
total_Lacz2_Lamina1 = 0
total_Lacz2_Lamina2 = 0

len_Lacz2 = 0
len_Lamina1 = 0
len_Lamina2 = 0

for name in chrnames:
    d_Lacz2 = tads_Lacz2.loc[tads_Lacz2.Chr == name]
    d_Lamina1 = tads_Lamin1.loc[tads_Lamin1.Chr == name]
    d_Lamina2 = tads_Lamin2.loc[tads_Lamin2.Chr == name]
    
    all_Lacz2 = np.append(d_Lacz2.Start.values, d_Lacz2.End.values)
    all_Lamina1 = np.append(d_Lamina1.Start.values, d_Lamina1.End.values)
    all_Lamina2 = np.append(d_Lamina2.Start.values, d_Lamina2.End.values)
    
    all_Lacz2 = np.unique(all_Lacz2)
    all_Lamina1 = np.unique(all_Lamina1)
    all_Lamina2 = np.unique(all_Lamina2)
    
    len_Lacz2 += len(all_Lacz2)
    len_Lamina1 += len(all_Lamina1)
    len_Lamina2 += len(all_Lamina2)
    
    Lacz2_Lamina1 = 0
    for b1 in all_Lacz2:
        for b2 in all_Lamina1:
            if b1 == b2 or b1+20000 == b2 or b1-20000 == b2:
                Lacz2_Lamina1+=1
                all_Lamina1 = np.delete(all_Lamina1, np.where(all_Lamina1 == b2))
                break
                
    Lacz2_Lamina2 = 0
    for b1 in all_Lacz2:
        for b2 in all_Lamina2:
            if b1 == b2 or b1+20000 == b2 or b1-20000 == b2:
                Lacz2_Lamina2+=1
                all_Lamina2 = np.delete(all_Lamina2, np.where(all_Lamina2 == b2))
                break
    
    total_Lacz2_Lamina1 += Lacz2_Lamina1
    total_Lacz2_Lamina2 += Lacz2_Lamina2
    
print('Total number of borders, Lacz2: ', len_Lacz2)
print('Total number of borders, Lamina1: ', len_Lamina1)
print('Total number of borders, Lamina2: ', len_Lamina2)
print(' ')
print('Shared borders, Lacz1 and Lamina1: ', total_Lacz2_Lamina1)
print('Shared borders, Lacz1 and Lamina2: ', total_Lacz2_Lamina2)
```
