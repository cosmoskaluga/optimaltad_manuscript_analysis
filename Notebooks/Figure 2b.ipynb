```r
library(ComplexHeatmap)
library(wesanderson)
library(circlize)
```

```r
ampl <- read.csv("../../Results/Lamina_2019/ChIPseq/No_Imputation_new/amplitudes.csv")
```

```r
colnames(ampl) <- c('Gamma', 'Control 1', 'Control 2', 'Lam-KD 2', 'Lam-KD 1')
```

```r
as.matrix(dist(t(ampl[, -c(1)])))
```

```r
col.pal <- colorRampPalette(c("white", "steelblue"))(30)
```

```r
ann_colors = list(
    Genotype = c(KO = wes_palette("Moonrise2")[1], 
                 WT = wes_palette("Moonrise2")[2]) 
)
```

```r
col_fun = colorRamp2(c(0, 9), c("white", "steelblue"))
```

```r
ph <- Heatmap(matrix = as.matrix(dist(t(ampl[, -c(1)]))), 
              name = "Euclidean distance",
              col = col_fun,
              rect_gp = gpar(col = "black", lwd = 0.8),
              show_row_names = T, 
              heatmap_legend_param = list(
                  legend_direction = "horizontal", 
                  legend_width = unit(4, "cm")),
             )
```
