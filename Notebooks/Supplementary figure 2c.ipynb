```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error
```

```python
prefix = "/Users/dmitriismirnov/Dropbox/Projects/optimalTAD/Results/Ing-Simmons/No_imputation_new/16kb/All_relicates_no_imputation/"
```

```python
amplitudes = pd.read_csv(prefix + 'amplitudes.csv')
```

```python
df1 = amplitudes[['Gamma','H3K27ac_R1']]
df2 = amplitudes[['Gamma','H3K27ac_R2']]

df1.columns = ['Gamma', 'Value']
df2.columns = ['Gamma', 'Value']
```

```python
df1['Replicate'] = 'H3K27ac Rep. 1'
df2['Replicate'] = 'H3K27ac Rep. 2'
```

```python
df_ampl = pd.concat([df1, df2])
```

```python
sns.set_theme(style="ticks")

fig, ax = plt.subplots(figsize=(7, 7))
sns.set_context("paper", font_scale=2.5)

my_pal = {"H3K27ac Rep. 1": "#2b8cbe", "H3K27ac Rep. 2": "#de2d26"}
sns.lineplot(data = df_ampl, x = "Gamma", 
             y = "Value", hue = "Replicate", 
            linewidth = 3, palette = my_pal)

sns.despine(offset=5, trim=False)
plt.ylabel('$\Delta h$ amplitude', fontsize = 25, labelpad = 2)
plt.xlabel(r'$\gamma$', fontsize = 25, labelpad = 2)
#plt.yticks(np.arange(-0.75, 1.26, 0.25))
#plt.ylim=(-0.85, 1.40)

plt.setp(ax.get_legend().get_texts(), fontsize='19') # for legend text
plt.setp(ax.get_legend().get_title(), fontsize='23') # for legend title
plt.grid()
```
